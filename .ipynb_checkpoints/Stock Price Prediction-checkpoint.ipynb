```python
future=model.make_future_dataframe(periods=365)
# # Create future DataFrame
# future_dates = pd.date_range(start='2022-01-01', periods=365, freq='D')
# future = pd.DataFrame({'ds': future_dates})
```

```python
# n_samples=365
# high_price = np.empty((n_samples, len(df['high_price'])))
# low_price  = np.empty((n_samples, len(df['low_price'])))
# close_price = np.empty((n_samples, len(df['close_price'])))
# open_price  = np.empty((n_samples, len(df['open_price'])))

# Generate the bootstrap samples
   
high_price = np.random.normal(loc=df['high_price'].mean(), scale=df['high_price'].std(), size=365)
low_price = np.random.normal(loc=df['low_price'].mean(), scale=df['low_price'].std(), size=365)
close_price = np.random.normal(loc=df['close_price'].mean(), scale=df['close_price'].std(), size=365)
open_price = np.random.normal(loc=df['open_price'].mean(), scale=df['open_price'].std(), size=365)
```

```python
future
future['high_price']=df['high_price']
future['low_price']=df['low_price']
future['close_price']=df['close_price']
future['open_price']=df['open_price']
# future['volume']=df['volume']
# future=future['high_price'].fillna(high_price)
# x=pd.Series(high_price)
```

```python
# Replace missing values in "high_price" column with values from "x" array
missing_indices = future['high_price'].index[future['high_price'].isna()]
future.loc[missing_indices, 'high_price'] = np.random.choice(high_price, size=len(missing_indices))

missing_indices = future['low_price'].index[future['low_price'].isna()]
future.loc[missing_indices, 'low_price'] = np.random.choice(low_price, size=len(missing_indices))

missing_indices = future['close_price'].index[future['close_price'].isna()]
future.loc[missing_indices, 'close_price'] = np.random.choice(close_price, size=len(missing_indices))

missing_indices = future['open_price'].index[future['open_price'].isna()]
future.loc[missing_indices, 'open_price'] = np.random.choice(open_price, size=len(missing_indices))

    
```

```python
# open column pre  
data_open=data.drop(columns=['High','Low','Close','Adj Close'])
data_heigh=data.drop(columns=['Open','Low','Close','Adj Close'])
data_Low=data.drop(columns=['High','Open','Close','Adj Close'])
data_close=data.drop(columns=['High','Open','Low','Adj Close'])
```

```python
# rename them for fbptophet model
data_open.rename(columns={'Date':'ds','Open':'y'},inplace=True)
data_heigh.rename(columns={'Date':'ds','High':'y'},inplace=True)
data_Low.rename(columns={'Date':'ds','Low':'y'},inplace=True)
data_close.rename(columns={'Date':'ds','Close':'y'},inplace=True)
```

```python
# applying fbprophet 
model_open= Prophet() 
model_heigh=Prophet() 
model_low= Prophet() 
model_close= Prophet() 
```

```python
#predict open data 
model_open.fit(data_open)
open_future=model_open.make_future_dataframe(356)
```

```python
#predict heigh data 
model_heigh.fit(data_heigh)
heigh_future=model_heigh.make_future_dataframe(356)
```

```python
#predict low data 
model_low.fit(data_Low)
low_future=model_low.make_future_dataframe(356)
```

```python
#predict close data 
model_close.fit(data_close)
close_future=model_close.make_future_dataframe(356)
```

```python
low_future.head()
```

```python
heigh_future.head()
```

```python
open_future.head()
```

```python
forcast_open=model_open.predict(open_future)
predicted_open=forcast_open['yhat'].tail(365)

forcast_low=model_low.predict(low_future)
predicted_low=forcast_low['yhat'].tail(365)

forcast_heigh=model_heigh.predict(heigh_future)
predicted_heigh=forcast_heigh['yhat'].tail(365)

forcast_close=model_close.predict(close_future)
predicted_close=forcast_heigh['yhat'].tail(365)

```

```python
model_heigh.plot(forcast_heigh)
model_low.plot(forcast_low)
model_open.plot(forcast_heigh)
model_open.plot(forcast_close)
plt.show()
```

```python
# plot both models in a single graph
fig, ax = plt.subplots(figsize=(10, 6))
model_heigh.plot_components(forcast_heigh)
model_low.plot_components(forcast_low)
model_open.plot_components(forcast_heigh)
model_open.plot_components(forcast_close)
plt.show()

```

```python

# plot all predictions in a single graph
fig, ax = plt.subplots(figsize=(10, 6))
model_heigh.plot(forcast_heigh, ax=ax )
model_low.plot(forcast_low, ax=ax )
model_open.plot(forcast_open, ax=ax)
model_close.plot(forcast_close, ax=ax)
lines = ax.lines
lines[0].set_color('blue')
lines[1].set_color('blue')
lines[2].set_color('green')
lines[3].set_color('green')
lines[4].set_color('yellow')
lines[5].set_color('yellow')
lines[6].set_color('red')
lines[7].set_color('red')
plt.title('Stocks Predictions')
plt.legend()
plt.show()
```

```python
# adding predicted regrassor values to our actual data

data_y=data.rename(columns={'Date':"ds",'Adj Close':'y'})
tasl_model=Prophet()
tasl_model.add_regressor("Open")
tasl_model.add_regressor("High")
tasl_model.add_regressor("Low")
tasl_model.add_regressor("Close")
tasl_model.fit(data_y)
tesla_future=tasl_model.make_future_dataframe(365)
```

```python
tesla_future
```

```python
final_Open__data=data_y['Open'].astype(str).append(predicted_open)
final_Close__data=data_y['Close'].astype(str).append(predicted_close)
final_High__data=data_y['High'].astype(str).append(predicted_heigh)
final_Low__data=data_y['Low'].astype(str).append(predicted_heigh)
```

```python
# check for dublicates and reindex them

final_Open__data=final_Open__data.reset_index(drop=True)
final_Close__data=final_Close__data.reset_index(drop=True)
final_High__data=final_High__data.reset_index(drop=True)
final_Low__data=final_Low__data.reset_index(drop=True)
```

```python

tesla_future['Open']=final_Open__data
tesla_future['High']=final_Close__data
tesla_future['Low']=final_High__data
tesla_future['Close']=final_Low__data
```

```python
tasla_prediciton=tasl_model.predict(tesla_future)
```
